```python
from keras.layers import Embedding,LSTM,Dense,SimpleRNN
from keras import models
from keras import layers
import numpy as np
import json
```

```python
model = models.Sequential()
model.add(layers.GRU(20))
model.add(layers.Dense(27,activation='softmax'))
```

```python
model_2 = models.Sequential()
model_2.add(layers.GRU(20))
model_2.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
model_2.compile(optimizer='rmsprop',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
obj_train=open("./test/json/sunny_train.json")
input_train=json.load(obj_train)
input_train=np.array(input_train)
```

```python

obj_output1=open("./test/json/sunny_output1.json")
output_train_1=json.load(obj_output1)
output_train_1=np.array(output_train_1)

note_train = np.zeros((len(output_train_1),27))
for i in range(len(output_train_1)):
    index = int(output_train_1[i]*2)
    note_train[i][index] = 1

time_train = np.zeros((len(output_train_1),10))
obj_output2=open("./test/json/sunny_output2.json")
output_train_2=json.load(obj_output2)
output_train_2=np.array(output_train_2)
for i in range(len(output_train_2)):
    index = int(output_train_2[i]/0.25 -1)
    time_train[i][index] = 1
    

print(time_train)
print(note_train)
```

```python
history=model.fit(input_train,note_train,batch_size=1,epochs=500)
```

```python
history=model_2.fit(input_train,time_train,batch_size=1,epochs=500)
```
